```python
import pandas as pd
import sqlite3
pd.set_option('display.max_rows', 500)
```

```python
list_hospitals = pd.read_csv("hospitals.csv")
```

```python
list_hospitals.head()
```

```python
len(list_hospitals)
```

```python
list_hospitals.columns
```

```python
list_hospitals.rename({'Open/Closed': 'Open_Closed',
                      'Local_Hospital_Network_(LHN)': 'Local_Hospital_Network_LHN',
                      'Primary_Health_Network_area_(PHN)': 'Primary_Health_Network_area',
                      'Remoteness_area_(code)': 'Remoteness area code',
                      'State_(abbrev)': 'State',
                      'Medicare_provider_no.': 'Medicare provider'}, axis=1, inplace=True)
```

```python
list_hospitals.columns
```

```python
#Create SQL Database from Dataframes

conn = sqlite3.connect('hospitals.db')


c = conn.cursor()
c.execute("CREATE TABLE IF NOT EXISTS hospital ({})".format(' ,'.join(list_hospitals.columns)))

for row in list_hospitals.iterrows():
    sql = "INSERT INTO hospital ({}) VALUES ({})".format(' ,'.join(list_hospitals.columns), ','.join(['?']*len(list_hospitals.columns)))
    c.execute(sql, tuple(row[1]))
conn.commit()
```

```python
cnx = sqlite3.connect('hospitals.db')

check = pd.read_sql("SELECT * FROM hospital", cnx)
check.head()
```
